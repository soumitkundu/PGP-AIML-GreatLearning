### Question 3.A:

```python
# Apply K-Means clustering for 2 to 10 clusters

# importing the required libraries for K-Means
from sklearn.cluster import KMeans
from sklearn.model_selection  import train_test_split
from scipy.stats import zscore
from scipy.spatial.distance import cdist
```

```python
# First lets assign the attributes to a dataframe
attr_car = df_car.iloc[:, 1:]
# now scaling the features before applying K-means
scaled_car = attr_car.apply(zscore)
```

```python
# importing warnings library to ignore unexpected warnings
import warnings
warnings.filterwarnings("ignore")

clusters = range(2, 10)
distortions = []

for k in clusters:
    model = KMeans(n_clusters = k)
    model.fit(scaled_car)
    prediction = model.predict(scaled_car)
    distortions.append(sum(np.min(cdist(scaled_car, model.cluster_centers_, 'euclidean'), axis = 1)) / scaled_car.shape[0])
```

### Question 3.B:

```python
# Plot a visual and find elbow point

plt.xlabel('k')
plt.ylabel('Mean distortion')
plt.title('Selecting different k values with Elbow Method')
plt.plot(clusters, distortions, 'bx-');
```

### Question 3.C:
- On the above visual, highlight which are the possible Elbow points

- The Elbow points are not very clear but we can go with K value 5.

### Question 3.D:

```python
# Train a K-means clustering model once again on the optimal number of clusters

# Trying with K value 5 as the elbow is more clear than others
optimal_model = KMeans(5)
optimal_model.fit(scaled_car)
prediction = optimal_model.predict(scaled_car)
```

### Question 1.B:

```python
# Check percentage of missing values and impute with correct approach
data_vehicle.isnull().sum() / data_car.shape[0] * 100
```

```python
def imputeMedianGroupByClass(df):
    for col in df.columns:
        if df[col].dtype.kind in 'iuf' and data_vehicle[col].isnull().sum() > 0:
#           'iucf': i = int, u = unsigned int, f = float
            df[col] = df[col].fillna(df.groupby('class')[col].transform('median'))
    return df

df_vehicle = imputeMedianGroupByClass(data_vehicle)
df_vehicle.isnull().sum()
```

### Question 1.C:

```python
# Visualize a Pie-chart and print percentage of values for variable ‘class’

class_values = list(df_vehicle['class'].value_counts())
class_keys = list(df_vehicle['class'].unique())
plt.pie(class_values, labels=class_keys, autopct='%1.1f%%');
```

### Question 2.A:

```python
# Split data into X and Y. [Train and Test optional]

df_vehicleX = df_vehicle.iloc[:, :-1]
df_vehicleY = df_vehicle['class']
```

### Question 2.B:

```python
# Standardize the Data

# import library to perform train test split
from sklearn.model_selection import train_test_split

# import Standardization library
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_vehicleX = pd.DataFrame(scaler.fit_transform(df_vehicleX))
scaled_vehicleX.columns = df_vehicleX.columns 

X_train, X_test, y_train, y_test = train_test_split(scaled_vehicleX, df_vehicleY, test_size = 0.20, random_state = 1)
X_train.head()
```

### Question 3.B:

### Question 3.C:

```python
# Apply PCA on the data with 10 components

# import PCA library
from sklearn.decomposition import PCA

pca = PCA(n_components = 10)
pca.fit(X_train)
```

### Question 3.D:

```python
# Visualize Cumulative Variance Explained with Number of Components
print("The Eigen Values")
print(pca.explained_variance_)
print("\n")
print("The Eigen Vectors")
print(pca.components_)
print("\n")
print("Percentage of variation explained by each Eigen Vector")
print(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize = (10 , 5))
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Number of Components')
plt.legend(loc = 'best')
# plt.tight_layout()
plt.bar(range(1, 11), pca.explained_variance_ratio_, alpha = 0.5, align = 'center', label = 'Individual explained variance')
plt.step(range(1, 11), np.cumsum(pca.explained_variance_ratio_), where='mid', label = 'Cumulative explained variance');
```

### Question 3.E:
