### Question 2.C:

```python
# Create a function that will accept a DataFrame as input and return pie-charts for all the appropriate Categorical features.
# Clearly show percentage distribution in the pie-chart.

# Libraries to handle data visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# library for confusion_matrix
from sklearn.metrics import confusion_matrix

# lets convert all the object datatype to categorical varisble
for col in churn_data.columns:
    if churn_data[col].dtype == 'object':
        churn_data[col] = pd.Categorical(churn_data[col])

# all the string datatype features converted to categorical
# now lets define the function for pie chart
def makePieChart(df):
    for feature in df.columns:
        if df[feature].dtype.name == 'category':
            feature_values = list(df[feature].value_counts())
            feature_keys = list(df[feature].unique())
            plt.title(feature)
            plt.axis('equal')
            plt.pie(feature_values, labels=feature_keys, autopct='%0.2f%%')
            plt.show();
            print("_"*100)
            print("\n")
            
makePieChart(churn_data)
```

### Question 2.E:

```python
# Encode all the appropriate Categorical features with the best suitable approach.

replaceData = {
    "Partner": { "No": 1, "Yes": 2 },
    "Dependents": { "No": 1, "Yes": 2 },
    "PhoneService": { "No": 1, "Yes": 2 },
    "PaperlessBilling": { "No": 1, "Yes": 2 },
    # "Contract": { "Month-to-month", 1, "One year": 2, "Two year": 3 },
    "Churn": {"No": 0, "Yes": 1 }
}
encodeOneHot = [ "gender", "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies", "Contract", "PaymentMethod" ]

churn_data=churn_data.replace(replaceData)
churn_data=pd.get_dummies(churn_data, columns=encodeOneHot)
churn_data.head()
```

```python
churn_data.info()
```

```python
# Still some columns are categorical type. lets change those to numetic
churn_data['Partner'] = pd.to_numeric(churn_data['Partner'])
churn_data['Dependents'] = pd.to_numeric(churn_data['Dependents'])
churn_data['PhoneService'] = pd.to_numeric(churn_data['PhoneService'])
churn_data['PaperlessBilling'] = pd.to_numeric(churn_data['PaperlessBilling'])
churn_data['Churn'] = pd.to_numeric(churn_data['Churn'])
```

```python
churn_data.info()
```

- Okay, now all the columns are nymeric only

### Question 2.F:

```python
# Split the data into 80% train and 20% test.

# import library to perform train test split
from sklearn.model_selection import train_test_split

churn_X = churn_data.drop("Churn", axis=1)
churn_y = churn_data["Churn"]
X_train, X_test, y_train, y_test = train_test_split(churn_X, churn_y, test_size = 0.20, random_state = 1)
churn_X.head()
```

### Question 2.G:

```python
# Normalize/Standardize the data with the best suitable approach.

# import Standardization library
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
churn_scaled_X = pd.DataFrame(scaler.fit_transform(churn_X))
churn_scaled_X.columns = churn_X.columns 

# lets split the scaled data again
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(churn_scaled_X, churn_y, test_size = 0.20, random_state = 1, stratify=churn_y)
churn_scaled_X.head()
```

### Question 3.A:

```python
# Train a model using Decision tree and check the performance of the model on train and test data

# import the library for DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# lets try the DecisionTreeClassifier model using default 'gini' criteria to split
dTree = DecisionTreeClassifier(criterion = 'gini', random_state = 1)
dTree.fit(X_scaled_train, y_train)
```

```python
# lets check the accuracy score of the model on train and test data
print("Performance on Training data =>", dTree.score(X_scaled_train, y_train))
print("Performance on Test data =>", dTree.score(X_scaled_test, y_test))
```

- We can clearly identify that the model is overfitting

```python
# lets check the max depth used by the model
dTree.tree_.max_depth
```

```python
# lets create another moder by reducing the overfinning
dTreeR = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, random_state = 1)
dTreeR.fit(X_scaled_train, y_train)
```

```python
# Now lets see the performance, whether the overfitting is reduced or not
print("Performance on Training data =>", dTreeR.score(X_scaled_train, y_train))
print("Performance on Test data =>", dTreeR.score(X_scaled_test, y_test))
```

- Now we can see the overfitting is reduced

### Question 3.B:

```python
# Use grid search and improve the performance of the Decision tree model, 

# Lets import the library for grid search
from sklearn.model_selection import GridSearchCV

# try some parameter for grid search
grid_param_DT = {
    'max_depth': range(1, 25),
    'criterion': [ 'gini', 'entropy' ]
}
grid_search = GridSearchCV(DecisionTreeClassifier(random_state = 1), grid_param_DT, cv = 5)
grid_search.fit(X_scaled_train, y_train)
```

```python
# lets see the best score, best parameters
print(grid_search.best_score_)
print(grid_search.best_params_)
```

```python
# check the performance of the model on train and test data, 
best_model_DT = grid_search.best_estimator_
best_model_DT.fit(X_scaled_train,y_train)
y_train_pred_DT = best_model_DT.predict(X_scaled_train)
y_test_pred_DT = best_model_DT.predict(X_scaled_test)
```

```python
print("Performance on Training data =>", best_model_DT.score(X_scaled_train, y_train))
print("Performance on Test data =>", best_model_DT.score(X_scaled_test, y_test))

# check the confusion_matrix
cm = confusion_matrix(y_test, y_test_pred_DT,labels=[0, 1])
df_cm = pd.DataFrame(cm, index = [i for i in ["No","Yes"]], columns = [i for i in ["No","Yes"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot = True, fmt = 'g');
```

```python
# provide the differences observed in performance in Q3.a and Q3.b
```

- we can see first the accuracy score was almost 100% on train and 72% on test data
- later it was improved to 79% and 77.6% respectively
- now the performance quite improved to 80% on train and 78.6% on test data
