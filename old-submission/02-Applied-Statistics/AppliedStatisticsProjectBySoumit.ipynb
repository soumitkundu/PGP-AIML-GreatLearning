### 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

```python
# Variable declaration
n = 10    # number of items in a sample
p = 0.05    # failure rate

k = np.arange(0,11)    # possible numbers of defective items

# using the binomial probability mass function(pmf) we can generate the probability distribution for each possible cases
binomial_pmf = stats.binom.pmf(k=k, n=n, p=p)

# using cumulative probability function(cdf) we can generate upto a certain range of distribution
binomial_cdf = stats.binom.cdf(k=k, n=n, p=p)
```

>    A. Probability that none of the items are defective?

### Ans:
Using the binomial function pmf we have generated the probability distribution for each possible cases in the variable binomial_pmf

```python
print("Probability that none of the items are defective is %1.4f" %binomial_pmf[0])
```

>    B. Probability that exactly one of the items is defective?

### Ans:
In the same variable binomial_pmf we already calculated the probability distribution for exactly one item

```python
print("Probability that exactly one of the items is defective is %1.4f" %binomial_pmf[1])
```

>    C. Probability that two or fewer of the items are defective?

### Ans:
Using the binomial function cdf we have generated the cumulative probability for upto each possible cases in the variable binomial_cdf

```python
print("Probability that two or fewer of the items are defective is %1.4f" %binomial_cdf[2])
```

>    D. Probability that three or more of the items are defective?

### Ans:
To findout three or more items defective we can simply substract the value of cumulative distribution of two from 1.

```python
print("Probability that three or more of the items are defective is %1.4f" %(1 - binomial_cdf[2]))
```

### 3. A car salesman sells on an average 3 cars per week.

- **Using Poisson law we can calculate the probablity.**

```python
rate = 3    # Average sales per week
n = np.arange(0,20)    # asuming that in a given week the salesman can sell as many as 19 (this number can be anything)

poisson_cdf = stats.poisson.cdf(n, rate)
```

>    A. What is Probability that in a given week he will sell some cars?

### Ans:

```python
print("Probability that in a given week he will sell some cars is %1.4f" %(1 - poisson_cdf[0]))
```

>    B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?

### Ans:

```python
print("Probability that in a given week he will sell 2 or more but less than 5 cars is %1.4f" %(poisson_cdf[4] - poisson_cdf[1]))
```

>    C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

### Ans:

```python
plt.plot(n,poisson_cdf, 'o-')
plt.title('Poisson distribution for cumulative probability')
plt.xlabel('Number of Cars sold per-week')
plt.ylabel('Cumulative probability of Cars sold per week')
plt.show()
```

### 4. Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed, marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours independently place orders with the same bot. Answer the following questions.

```python
# Variable declaration
n = 3    # number of orders
p = 0.868    # success rate

k = np.arange(0,4)    # possible numbers of successful orders

# using the binomial pmf function we can generate the probability distribution for each possible cases
binomial_pmf = stats.binom.pmf(k=k, n=n, p=p)

# using binomial cdf function we can generate the cumulative distribution of upto a certain range
binomial_cdf = stats.binom.cdf(k=k, n=n, p=p)
```

>    A. What is the probability that all three orders will be recognised correctly?

### Ans:

Using binomial probability mass function we have calculated the distribution of each possible cases.

```python
print('The probability that all three orders will be recognised correctly is %1.4f' %binomial_pmf[3])
```

>    B. What is the probability that none of the three orders will be recognised correctly?

### Ans:

Using the same pmf function we already calculated the possible rate that none will be correct

```python
print('The probability that none of the three orders will be recognised correctly is %1.4f' %binomial_pmf[0])
```

> C. What is the probability that at least two of the three orders will be recognised correctly?

### Ans:

Using the binomial cdf function we have generated the cumulative probability for upto each possible cases. Now to calculate at least two success case we need to find out 1 - probability of upto 1 failure case.

```python
print('The probability that at least two of the three orders will be recognised correctly is %1.4f' %(1 - binomial_cdf[1]))
```
