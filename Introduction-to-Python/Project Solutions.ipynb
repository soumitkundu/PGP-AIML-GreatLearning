```python
# Import Required Modules
# string to get all letters and digits
import string

# for random operation
import random

# for date and time operation
from datetime import datetime

# numpy and pandas for numbers and DataFrame operation
import numpy as np
import pandas as pd
```

> 1. Create a pandas data frame consisting of 4 columns and 20 rows.
> 2. Column 1 should have values between 0 to 25.

```python
col_1 = pd.Series(np.random.randint(0, 25, 20))
col_1
```

> 3. Column 2 should have values between -10 to 10.

```python
col_2 = pd.Series(np.random.randint(-10, 10, 20))
col_2
```

> 4. Column 3 should have values by multiplying column 1 and 2.

```python
col_3 = pd.Series(col_1 * col_2)
col_3
```

```python
df = pd.DataFrame({"col_1": col_1, "col_2": col_2, "col_3": col_3})
df
```

> 5. Column 4 should have value “Positive” if column 3 is a positive number.
> 6. Column 4 should have value “Negative” if column 3 is a negative number.

```python
isPositiveOrNegative = lambda num: "Positive" if num > 0 else "Negative" if num < 0 else None
df["col_4"] = df["col_3"].apply(isPositiveOrNegative)
df
```

> 7. Create a new data frame of 1 column and 20 rows having random alphanumeric values of size 5. Name the column as ‘Password’

```python
def genedatePassword(str_len):
    source_str = string.ascii_letters + string.digits
    rand_str = ""
    for i in range(0, str_len):
        rand_str += source_str[random.randint(0, (len(source_str) - 1))]
    
    return rand_str
```

```python
df_2 = pd.DataFrame(data = np.full(20, 5), columns = ["Password"])
df_2 = df_2.astype('object')
df_2["Password"] = df_2["Password"].apply(genedatePassword)
df_2
```

> 8. Merge both the data frames to one single data frame.

```python
df = pd.concat([df, df_2], axis = 1)
df
```

> 9. Create a 2 new columns i.e. ‘Word’ and ‘Number’ by separating the alphabets and numbers from column “Password”.

```python
def extractAlphabetsOrNumbers(str, return_type):
    return_val = ""
    for ch in str:
        if return_type == "Word" and ch.isalpha():
            return_val += ch
        elif return_type == "Number" and ch.isdigit():
            return_val += ch
        
    return return_val
```

```python
df["Word"] = df.apply(lambda x: extractAlphabetsOrNumbers(x["Password"], "Word"), axis = 1)
df["Number"] = df.apply(lambda x: extractAlphabetsOrNumbers(x["Password"], "Number"), axis = 1)
df
```

> 10. Create a new column ‘Anomaly’. This column should have value ‘Found’ if the row entry in ‘Number’ is empty.

```python
df["Anomaly"] = df.apply(lambda x: "Found" if x["Number"] == "" else "", axis = 1)
df
```

> 11. Insert a new column “Date” by inserting the current date. [ format: mmm-dd-yyyy ]

```python
dt_now = datetime.now()
df["Date"] = df.apply(lambda x: dt_now.strftime("%b-%d-%Y"), axis = 1)
df
```

> 12. Insert a new column “Time” by inserting the current time. [ format: hh-mm-ss ]

```python
dt_now = datetime.now()
df["Time"] = df.apply(lambda x: dt_now.strftime("%H-%M-%S"), axis = 1)
df
```

> 13. Create a new column ‘Comment’. This column should have value ‘Non office hours of working’ if the row entry in ‘Time’ is not between 9 am to 7 pm.

```python
df["Comment"] = df.apply(lambda x: "Non office hours of working" if int(x["Time"][:2]) < 9 or int(x["Time"][:2]) > 18 else "", axis = 1)
df
```

> 14. Export the data frames to excel, csv, html and json formats.

```python
# excel
df.to_excel('df.xlsx', index = False)
# csv
df.to_csv('df.csv', index = False)
# html
df.to_html('df.html', index = False)
# json
df.to_json('df.json', orient="records")
```
